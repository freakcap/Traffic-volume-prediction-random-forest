```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import datetime
import sklearn.metrics
from sklearn import preprocessing
from tpot import TPOTRegressor
sns.set()
```

```python
from datetime import datetime
train['date_time'] = pd.to_datetime(train.date_time)
train['year'] = train['date_time'].dt.year
train['month'] = train['date_time'].dt.month
train['day'] = train['date_time'].dt.day
train['dayofyear'] = train['date_time'].dt.dayofyear
train['weekofyear'] = train['date_time'].dt.weekofyear
train['dayofweek'] = train['date_time'].dt.dayofweek
train['hour'] = train['date_time'].dt.hour
train.head()
```

```python
vard = 0
varm = 0
var = ""
for index,row in train.iterrows():
    if(row['is_holiday'] != "None"):
        vard = row['day']
        varm = row['month']
        var = row['is_holiday']
    if(row['day'] == vard and row['month'] == varm):
        train.loc[index,'is_holiday']=var
```

```python
train['is_holiday'].value_counts()
```

```python
for index, row in train.iterrows():
    if(row['temperature']==0):
        train.drop(index,inplace=True)
    if(row['rain_p_h']>30):
        train.drop(index,inplace=True)
    if(row['snow_p_h']>0.2):
        train.drop(index,inplace=True)
```

```python
train.drop(['air_pollution_index'], axis=1, inplace=True)
# train.drop(['visibility_in_miles'], axis=1, inplace=True)
train.drop(['wind_direction'], axis=1, inplace=True)
# train.drop(['clouds_all'], axis=1, inplace=True)
# train.drop(['weather_description'], axis=1, inplace=True)
# train.drop(['weather_type'], axis=1, inplace=True)
# train.drop(['snow_p_h'], axis=1, inplace=True)
# train.drop(['rain_p_h'], axis=1, inplace=True)
# train.drop(['humidity'], axis=1, inplace=True)
train.drop(['dew_point'], axis=1, inplace=True)
# train.drop(['wind_speed'], axis=1, inplace=True)
# train.drop(['temperature'], axis=1, inplace=True)
# train.drop(['is_holiday'], axis=1, inplace=True)
# train.drop(['year'], axis=1, inplace=True)
# train.drop(['day'], axis=1, inplace=True)
train.head()
```

```python
train['peak_time'] = 0
```

```python
for index,row in train.iterrows():
    if(row['hour'] >= 22 and row['hour'] <6):
        train.loc[index,'peak_time'] = 0
    elif((row['hour'] >= 19 and row['hour'] <22)):
        train.loc[index,'peak_time'] = 2
    elif((row['hour'] > 14 and row['hour'] <19)):
        train.loc[index,'peak_time'] = 9
    elif((row['hour'] >= 6 and row['hour'] <=9)):
        train.loc[index,'peak_time'] = 1
    else:
        train.loc[index,'peak_time'] = 1
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,OneHotEncoder
labelencoder = LabelEncoder()
train['weather_type'] = labelencoder.fit_transform(train['weather_type'])
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,OneHotEncoder
labelencoder = LabelEncoder()
train['weather_description'] = labelencoder.fit_transform(train['weather_description'])
train.head()
```

```python
train.corr()
```

```python
train.drop(['date_time'], axis=1, inplace=True)
train.head()
```

```python

y = train["traffic_volume"]

X = train.drop(["traffic_volume"],axis=1)

X.shape
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split 

x_train , x_test, y_train, y_test = train_test_split(X,y,test_size=0.001,random_state=0)
```

```python
scaler=MinMaxScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
model = RandomForestRegressor(n_jobs=-1)
```

```python
estimators = np.arange(10, 500, 10)
scores = []
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(X_train, y_train)
    scores.append(model.score(X_test, y_test))
plt.title("Effect of n_estimators")
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```

```python
scores
```

```python
lr = RandomForestRegressor(bootstrap=True,n_estimators=100) 
```

```python
lr.fit(X_train,y_train)
```

```python
y_predictions = lr.predict(X_test)
r2 = sklearn.metrics.r2_score(y_test, y_predictions)
mae = sklearn.metrics.mean_absolute_error(y_test, y_predictions)
mse = sklearn.metrics.mean_squared_error(y_test, y_predictions)
lmse = sklearn.metrics.mean_squared_log_error(y_test, y_predictions)
rmse = np.sqrt(mse)
lrmse = np.sqrt(lmse)
print("R2 score:", r2)
print("MAE:", mae)
print("MSE:", mse)
print("LMSE:", lmse)
print("RMSE:", rmse)
print("LRMSE:", lrmse)
```
